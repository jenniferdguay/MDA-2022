```python
import pandas as pd
```

```python
# read in csv files
state_code = pd.read_csv("State code.csv")
pop_est = pd.read_csv("nst-est2019-alldata.csv")
pop_est21 = pd.read_excel("NST-EST2021-POP.xlsx", header = 3)
precip17 = pd.read_csv("precip-201707-ytd.csv", header = 3)
precip18 = pd.read_csv("precip-201807-ytd.csv", header = 3)
precip21 = pd.read_csv("precip-202107-ytd.csv", header = 3)
temp17 = pd.read_csv("tempav-201707-ytd.csv", header = 3)
temp18 = pd.read_csv("tempav-201807-ytd.csv", header = 3)
temp21 = pd.read_csv("tempav-202107-ytd.csv", header = 3)

print(state_code.head())
print(pop_est.tail())
#print(pop_est21.head())
#rint(temp17.head())
```

```python
## Merge tables

state_code.drop('Abbrev', axis = 1, inplace=True)

# population '17 & '18
pop_est = pop_est.loc[:,['NAME', 'POPESTIMATE2017', 'POPESTIMATE2018']]
pop_est.rename(columns={'NAME':'State', 'POPESTIMATE2017':'Pop17', 'POPESTIMATE2018':'Pop18'}, inplace=True)
state_info = pd.merge(state_code, pop_est, how='left',on = 'State')

# population '21
pop_est21.rename(columns={'Unnamed: 0':'State', 2021:'Pop21'}, inplace=True)
pop_est21.drop(['Unnamed: 1', 2020], axis = 1, inplace=True)
pop_est21.State = pop_est21.State.str.replace('.','', regex=True)
state_info = state_info.merge(pop_est21, how='left',on='State')

# Precipitation
precip17.rename(columns={'Location':'State','Value':'Precip17'}, inplace=True)
precip17 = precip17[['State','Precip17']]
state_info = state_info.merge(precip17, how='left',on='State')
precip18.rename(columns={'Location':'State','Value':'Precip18'}, inplace=True)
precip18 = precip18[['State','Precip18']]
state_info = state_info.merge(precip18, how='left',on='State')
precip21.rename(columns={'Location':'State','Value':'Precip21'}, inplace=True)
precip21 = precip21[['State','Precip21']]
state_info = state_info.merge(precip21, how='left',on='State')

# Temperature
temp17.rename(columns={'Location':'State','Value':'Temp17'}, inplace=True)
temp17 = temp17[['State','Temp17']]
state_info = state_info.merge(temp17, how='left',on='State')
temp18.rename(columns={'Location':'State','Value':'Temp18'}, inplace=True)
temp18 = temp18[['State','Temp18']]
state_info = state_info.merge(temp18, how='left',on='State')
temp21.rename(columns={'Location':'State','Value':'Temp21'}, inplace=True)
temp21 = temp21[['State','Temp21']]
state_info = state_info.merge(temp21, how='left',on='State')
state_info.head()
```

```python
# Check for missing values
print(state_info.isnull().any())
print(state_info.isnull().sum())
# Find states with missing values
print(state_info.isnull().sum(axis = 1))
```
