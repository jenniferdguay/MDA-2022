## Creating the states and city columns 

Because our aim is to analyse state wide issues on water supply, we will create a column with the names of the states for each city. 

### 2021 US dataset 

The 2021 data set has the names of the state and city in the column **Account Name**, therefore we will seperate these into two seperate columns. 

`````python
#create the variable City and State
usdf_21[['City', 'State']] = usdf_21["Account Name"].str.split(',', expand=True)

#drop account Name
usdf_21.drop(['Account Name'], axis = 1, inplace=True)

usdf_21.head(5)
`````

### 2018 US dataset

On the other hand, the 2018 and 2017 datasets do not have any information on the states of the cities. Additionally, since many US cities such as Atlanta exist in multiple states we cannot simply merge a city from our datasets to a state dataset found online. However, because the 2018 and 2017 datasets do include the cities' coordinates, we can use this information to find the states using the library ````GEOPY````.

`````python
#import geopy for finding states
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim
import pandas as pd
geolocator = Nominatim(user_agent="user_agent")

# add rate limit to send 1 request per second
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
`````

`````python
# The same is then done for the 2018 dataset
usdf_18 = df_2018[['Country', 
            'Account Name',
            'Water security risk drivers', 
            'Anticipated timescale',
            'Estimated magnitude of potential impact',
            'Risk description',"City Location"]].copy()


#add the condition that country is USA and the water security risk is a current problem 
usdf_18 = usdf_18[(usdf_18.Country == "United States of America") & (usdf_18['Anticipated timescale'] == "Current")]
usdf_18["Year"] = 2018
`````

`````python
#Creating a coordinateds tuple for creating the state variable using the city location feature

#clean the coorndinates string
usdf_18['City Location'] = usdf_18['City Location'].str.replace(r'[()]',"")
usdf_18['City Location'] = usdf_18['City Location'].str.replace('_', '', regex=False)
usdf_18[['lat','lng']]=usdf_18["City Location"].str.split(',',expand=True)



#creating a coordinate tuple
usdf_18['coordinates'] = usdf_18[['lat', 'lng']].apply(tuple, axis=1)
`````

`````python
state_list = [] # Initialize empty list

# get location coordinates
geo_locv  = usdf_18.coordinates.values

#manually add Hawaii coordinates because the api doesn't recognize the ones given by dataset
geo_locv[16] = (21.304547,-157.855676)

for cord in geo_locv:
    # send request
    location = reverse(cord, addressdetails=True)
    # get state value
    state = location.raw["address"].get("state")
    # store state value
    state_list.append(state)
# assign back states
usdf_18['State'] = state_list
`````

`````python
#Getting state abbreviations 
usdf_18 = pd.merge(usdf_18,state_df,on='State')

#dropping unwanted columns
usdf_18.drop(['City Location', 'lat','lng','coordinates','State','Abbrev'], axis=1,inplace=True)

#Renaming Code to State
usdf_18.rename(columns={"Code": "State"},inplace=True)

#Creating a city variable
usdf_18['City'] = usdf_18['Account Name'].str.split(',').str[0]

#dropping account variable
usdf_18.drop(['Account Name'], axis = 1, inplace=True)
`````

`````python
#Let's check that we have the state and cities seperately 
usdf_18.head(5)
`````

### 2017 US dataset

We repeat the same procedure as in the 2018 dataset for the 2017 dataset to create the state and city columns. 

`````python
#Creating a coordinate tuple for finding the state using the city location feature

#clean the coorndinates string
usdf_17['City location'] = usdf_17['City location'].str.replace(r'[()]',"")
usdf_17['City location'] = usdf_17['City location'].str.replace('_', '', regex=False)
usdf_17[['lat','lng']]=usdf_17["City location"].str.split(',',expand=True)

#creating the coordinate tuple
usdf_17['coordinates'] = usdf_17[['lat', 'lng']].apply(tuple, axis=1)
`````

`````python
state_list = [] # Initialize empty list

# get location coordinates
geo_locv  = usdf_17.coordinates.values
for cord in geo_locv:
    # send request
    location = reverse(cord, addressdetails=True)
    # get state value
    state = location.raw["address"].get("state")
    # store state value
    state_list.append(state)
# assign back states
usdf_17['State'] = state_list
`````

`````python
#Getting state abbreviations 
usdf_17 = pd.merge(usdf_17,state_df,on='State')

#dropping unwanted columns
usdf_17.drop(['City location', 'lat','lng','coordinates','State','Abbrev'], axis=1, inplace=True)

#Renaming Code to State
usdf_17.rename(columns={"Code": "State"},inplace=True)

#Creating a city variable
usdf_17['City'] = usdf_17['Account Name'].str.split(',').str[0]

#dropping account variable
usdf_17.drop(['Account Name'], axis = 1, inplace=True)
`````

`````python
#Let's check that we have the state and cities seperately 
usdf_17.head(5)
`````

Having created our US dataset, we might want to check for missing values and values spelled differently. 

`````python
#check the values for the waster security risk drivers
usdf["Water security risk drivers"].unique()
`````

`````python
#check the values for estimated magnitiude of potential impact
usdf["Estimated magnitude of potential impact"].unique()
`````

The variable for **waste security risk drivers** and **estimated magnitude of potential** impact have a null element. Additionally **estimate magnitude of potential** impact have two values of *less serious* written differently (i.e., one as "Less Serious" and the other as "Less serious"). Moreover **waste security risk drivers** has "Increased water stress" and "Increased water scarcity" sometimes written as one value "Increased water stress or scarcity", as well as "Inadequate or ageing water supply infrastructure" sometimes written as "Inadequate or ageing infrastructure" or "Inadequate or aging infrastructure". 

Therefore, we will remove rows with null elements and modify the writing of the erroneous values so that they match up. 

`````python
#remove all the rows with nan 
usdf.dropna(axis = 0, inplace=True)

#replace "Less Serious by "Less serious"
usdf["Estimated magnitude of potential impact"].replace("Less Serious","Less serious", inplace=True)

#replace "Increased water stress" to "Increased water stress or scarcity"
usdf["Water security risk drivers"].replace("Increased water stress","Increased water stress or scarcity", inplace=True)

#replace "Increased water scarcity" to "Increased water stress or scarcity"
usdf["Water security risk drivers"].replace("Increased water scarcity","Increased water stress or scarcity", inplace=True)

#replace "Inadequate or aging infrastructure" to "Inadequate or ageing water supply infrastructure"
usdf["Water security risk drivers"].replace("Inadequate or aging infrastructure","Inadequate or ageing water supply infrastructure", inplace=True)

#replace "Inadequate or ageing infrastructure" to "Inadequate or ageing water supply infrastructure"
usdf["Water security risk drivers"].replace("Inadequate or ageing infrastructure","Inadequate or ageing water supply infrastructure", inplace=True)
`````

`````python
#Let's check for the presence of null elements in each column
usdf.isna().sum()
`````

`````python
#Let's check that the variable waste security risk drivers and estimated magnitude of potential impact no longer have a nan
print("Values of Water security risk drivers: {}.\n".format(usdf["Water security risk drivers"].unique()))
print("Values of Estimated magnitude of potential impact: {}.".format(usdf["Estimated magnitude of potential impact"].unique()))
`````

Great, looks like missing values have been removed and values have been fixed! Let's now drop the **Country** because we do not need it anymore, as well as **Anticipated timescale** because they were all chosen to be "Current".
